```python
import glob
import xml.etree.ElementTree as ET
import collections
import time
import pandas as pd
import re
```

```python
list_of_files = glob.glob('../data/clinicaltrials_xml/*/*' + '/*.xml')
listOflist = []
```

```python
counter = 0
for input_file in list_of_files:
    dfList = []
    extracted_data = {}
    tree = ET.parse(input_file)
    root = tree.getroot()
    
    try:
        nct_id = root.find('id_info').find('nct_id').text
#         print("Nct:", nct_id)
        dfList.append(nct_id)
    except:
        dfList.append(None)
    
    try:
        brief_title = root.find('brief_title').text
        brief_title = brief_title.replace("\n", " ")
        re.sub(' +', ' ', brief_title)
        brief_title = ' '.join(brief_title.split())
#         print("brief_title:", brief_title)
        dfList.append(brief_title.strip())
    except:
        dfList.append(None)
    
    try:
        official_title = root.find('official_title').text
        official_title = official_title.replace("\n", " ")
        re.sub(' +', ' ', official_title)
        official_title = ' '.join(official_title.split())
#         print("brief_title:", official_title)
        dfList.append(official_title.strip())
    except:
        dfList.append(None)
    
    try:
        brief_summary = root.find('brief_summary').find('textblock').text
        brief_summary = brief_summary.replace("\n", " ")
        re.sub(' +', ' ', brief_summary)
        brief_summary = ' '.join(brief_summary.split())
#         print("brief_summary:", brief_summary.strip())
        dfList.append(brief_summary.strip())
    except:
        dfList.append(None)
    
    
    try:
        detailed_description = root.find('detailed_description').find('textblock').text
        detailed_description = detailed_description.replace("\n", " ")
        re.sub(' +', ' ', detailed_description)
        detailed_description = ' '.join(detailed_description.split())
#         print("detailed_description:", detailed_description)
        dfList.append(detailed_description.strip())
    except:
        dfList.append(None)
    
    #eligibility
    try:
        eligibility = root.find('eligibility').find('criteria').find('textblock').text
        eligibility = eligibility.replace("\n", " ")
        re.sub(' +', ' ', eligibility)
        eligibility = ' '.join(eligibility.split())
        extracted_data['eligibility'] = eligibility
        
    except:
        extracted_data['eligibility'] = None
    dfList.append(extracted_data['eligibility'])
    
    
    try:
        overall_status = root.find('overall_status').text
        overall_status = overall_status.replace("\n", " ")
        re.sub(' +', ' ', overall_status)
        overall_status = ' '.join(overall_status.split())
#         print("overall_status:", overall_status)
        dfList.append(overall_status)
    except:
        dfList.append(None)
    
    try:
        condition = root.find('condition').text
        condition = condtion.replace("\n", " ")
        re.sub(' +', ' ', condition)
        condition = ' '.join(condition.split())
#         print("condition:", condition)
        dfList.append(condition)
    except:
        dfList.append(None)
    
    try:
        gender = root.find('eligibility').find('gender').text
        gender.replace("\n", " ")
        re.sub(' +', ' ', gender)
        gender = ' '.join(gender.split())
#         print("gender:", gender)
        dfList.append(gender)
    except:
        dfList.append(None)
    
    
    #minimum_age
    try:
        minimum_age = root.find('eligibility').find('minimum_age').text
        try:
            extracted_data['minimum_age'] = int(minimum_age.split(' ')[0])
#             print("minimum_age:", extracted_data['minimum_age'])
        except:
            extracted_data['minimum_age'] = 0
    except:
        extracted_data['minimum_age'] = None
    dfList.append(extracted_data['minimum_age'])

    #maximum_age
    try:
        maximum_age = root.find('eligibility').find('maximum_age').text
        try:
            extracted_data['maximum_age'] = int(maximum_age.split(' ')[0])
#             print("maximumAge:", extracted_data['maximum_age'])
        except:
            extracted_data['maximum_age'] = 99
    except:
        extracted_data['maximum_age'] = None
    dfList.append(extracted_data['maximum_age'])
    
    try:
        mesh_term_list = []
        mesh_term = root.find('condition_browse').findall('mesh_term')
        for index, item in enumerate(mesh_term):
            mesh_term_list.append(item.text)
        dfList.append(";".join(mesh_term_list))
    except:
        dfList.append(None)
    
    try:
        keyword_list = []
        keyword = root.findall('keyword')
        for index, item in enumerate(keyword):
            keyword_list.append(item.text)
        dfList.append(";".join(keyword_list))  
    except:
        dfList.append(None)
    
#     print(input_file, nct_id)
    listOflist.append(dfList)
    counter += 1
    
#     if counter == 1000:
#         break
```

```python
df = pd.DataFrame.from_records(listOflist, columns=['nct_id', 'brief_title', 'official_title', 'brief_summary', 'detailed_description', 'eligibility', 'overall_status', 'condition', 'gender', 'minimum_age', 'maximum_age', 'mesh_term_list', 'keyword_list'])
```
