```python
def dcg(rel_list):
	rank = 1
	score = 0
	for item in rel_list:
		score += (pow(2, item)-1)/(math.log(rank+1, 2))
		rank += 1
	return score
```

```python
def cal_precision(relevance):
    count = 0
    precisionList = []
    
    #P@5:
    if len(relevance) >= 5:
        for rel in relevance[:5]:
            if rel:
                count += 1
        precisionList.append(round(count/5, 2))
    else:
        precisionList.append("-")
        
    #P@10:
    if len(relevance) >= 10:
        for rel in relevance[5:10]:
            if rel:
                count += 1
        precisionList.append(round(count/10, 2))
    else:
        precisionList.append("-")
    
    #P@15:
    if len(relevance) >= 15:
        for rel in relevance[10:15]:
            if rel:
                count += 1
        precisionList.append(round(count/15, 2))
    else:
        precisionList.append("-")
    
    
    #P@20:
    if len(relevance) >= 20:
        for rel in relevance[15:20]:
            if rel:
                count += 1
        precisionList.append(round(count/20, 2))
    else:
        precisionList.append("-")
        
    return precisionList
```

```python
def addZeroInsteadDash(precisionList):
    newL = []
    for item in precisionList:
        if item == '-':
            newL.append(0.0)
        else:
            newL.append(float(item))
    return newL
```

```python
def plotFigure(beforePrecision, afterPrecision, title):
    plt.title(title)
    plt.xlabel("Queries")
    plt.ylabel("Precision Values")
    
    plt.plot(beforePrecision, 'ro')
    plt.plot(afterPrecision, 'go')
    plt.ylim((0,1))
    plt.show()
```
